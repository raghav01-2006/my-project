```python
import pandas as pd

# 1. Create a Pandas Series from Dictionary

data_dict = {
    "Math": 85,
    "Science": 90,
    "English": 78
}

series_from_dict = pd.Series(data_dict)

print("Series from Dictionary:")
print(series_from_dict)


# 2. Create a Pandas Series from List

data_list = [10, 20, 30, 40, 50]

series_from_list = pd.Series(data_list)

print("\nSeries from List:")
print(series_from_list)


# 3. Access Elements of a Series

print("\nAccessing Elements:")
print("First Element:", series_from_list[0])
print("Elements from index 1 to 3:")
print(series_from_list[1:4])
print("Science Marks:", series_from_dict["Science"])
```

```python
import pandas as pd

# 1. Make a Pandas DataFrame with 2D Python List

data_2d = [
    [1, "Yogesh", 18],
    [2, "Rahul", 19],
    [3, "Aman", 20]
]

df_2d = pd.DataFrame(data_2d, columns=["ID", "Name", "Age"])

print("DataFrame from 2D List:")
print(df_2d)


# 2. Create DataFrame from Python Dictionary

data_dict = {
    "Name": ["Yogesh", "Rahul", "Aman"],
    "Age": [18, 19, 20],
    "City": ["Jaipur", "Delhi", "Mumbai"]
}

df_dict = pd.DataFrame(data_dict)

print("\nDataFrame from Dictionary:")
print(df_dict)


# 3. Create DataFrame using List of Lists

list_of_lists = [
    ["Laptop", 50000],
    ["Mobile", 20000],
    ["Tablet", 15000]
]

df_list = pd.DataFrame(list_of_lists, columns=["Product", "Price"])

print("\nDataFrame from List of Lists:")
print(df_list)


# 4. Create DataFrame using List of Tuples

list_of_tuples = [
    ("Math", 85),
    ("Science", 90),
    ("English", 78)
]

df_tuple = pd.DataFrame(list_of_tuples, columns=["Subject", "Marks"])

print("\nDataFrame from List of Tuples:")
print(df_tuple)


# 5. Create DataFrame from List of Dictionaries

list_of_dicts = [
    {"Name": "Yogesh", "Age": 18, "City": "Jaipur"},
    {"Name": "Rahul", "Age": 19, "City": "Delhi"},
    {"Name": "Aman", "Age": 20, "City": "Mumbai"}
]

df_dicts = pd.DataFrame(list_of_dicts)

print("\nDataFrame from List of Dictionaries:")
print(df_dicts)
```

```python
import pandas as pd

data = {
    "ID": [1, 2, 3, 4],
    "Name": ["Yogesh", "Rahul", "Aman", "Rohit"],
    "Age": [18, 22, 19, 25],
    "Marks": [85, 90, 78, 88]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)


# 1. Different ways to iterate over rows

print("\nUsing iterrows():")
for index, row in df.iterrows():
    print(index, row["Name"], row["Marks"])


print("\nUsing itertuples():")
for row in df.itertuples():
    print(row.ID, row.Name, row.Age)


# 2. Selecting rows based on conditions

print("\nStudents with Marks greater than 80:")
print(df[df["Marks"] > 80])


# 3. Select any row using iloc[]

print("\nSelect 2nd row using iloc:")
print(df.iloc[1])


# 4. Limited rows selection with given column

print("\nSelect first 2 rows with Name and Marks columns:")
print(df.loc[0:1, ["Name", "Marks"]])


# 5. Drop rows based on condition

print("\nDrop rows where Age > 20:")

df_dropped = df[df["Age"] <= 20]

print(df_dropped)


# 6. Insert row at given position

new_row = {"ID": 5, "Name": "Karan", "Age": 21, "Marks": 92}

# Split dataframe and insert row at position 2
top = df.iloc[:2]
bottom = df.iloc[2:]

new_df = pd.concat([top, pd.DataFrame([new_row]), bottom]).reset_index(drop=True)

print("\nDataFrame after inserting new row at position 2:")
print(new_df)


# 7. Create list from rows in DataFrame

print("\nCreate list from DataFrame rows:")

row_list = df.values.tolist()

print(row_list)
```
