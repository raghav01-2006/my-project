```python
import csv
name1 = input("Enter Name 1: ")
address1 = input("Enter Address 1: ")
mobile1 = input("Enter Mobile 1: ")
email1 = input("Enter Email 1: ")

name2 = input("Enter Name 2: ")
address2 = input("Enter Address 2: ")
mobile2 = input("Enter Mobile 2: ")
email2 = input("Enter Email 2: ")
with open("address_book.csv", "w", newline='') as file:
    writer = csv.writer(file)

    writer.writerow(["Name", "Address", "Mobile", "Email"])

 
    writer.writerow([name1, address1, mobile1, email1])
    writer.writerow([name2, address2, mobile2, email2])

print("CSV File Created Successfully!")
```

```python
import sqlite3
conn = sqlite3.connect("college.db")
cur = conn.cursor()
cur.execute("""
CREATE TABLE IF NOT EXISTS Student(
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    city TEXT
)
""")
cur.execute("""
CREATE TABLE IF NOT EXISTS Course(
    course_id INTEGER PRIMARY KEY,
    course_name TEXT,
    fees INTEGER
)
""")

print("Tables Created Successfully")
```

```python
cur.execute("INSERT INTO Student VALUES(1, 'Rahul', 20, 'Delhi')")
cur.execute("INSERT INTO Student VALUES(2, 'Aman', 21, 'Mumbai')")
cur.execute("INSERT INTO Student VALUES(3, 'Neha', 19, 'Pune')")

cur.execute("INSERT INTO Course VALUES(101, 'Python', 5000)")
cur.execute("INSERT INTO Course VALUES(102, 'Java', 6000)")

conn.commit()

print("Records Inserted Successfully")
```

```python
cur.execute("SELECT * FROM Student")

rows = cur.fetchall()

for row in rows:
    print(row)
```

```python
cur.execute("UPDATE Student SET city='Jaipur' WHERE id=1")

conn.commit()

print("Record Updated Successfully")
```
