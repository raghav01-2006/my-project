```python
name = input("Enter student name: ")
student_class = input("Enter class: ")

mark1 = float(input("Enter marks for Subject 1: "))
mark2 = float(input("Enter marks for Subject 2: "))
mark3 = float(input("Enter marks for Subject 3: "))
mark4 = float(input("Enter marks for Subject 4: "))
mark5 = float(input("Enter marks for Subject 5: "))

total = mark1 + mark2 + mark3 + mark4 + mark5
percentage = (total / 500) * 100

print("\n----- Student Result -----")
print("Name:", name)
print("Class:", student_class)
print("Total Marks:", total)
print("Percentage:", percentage, "%")
```

```python
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

result = str1 + str2

print("Concatenated String:", result)

print("lower():", result.lower())
print("upper():", result.upper())
print("title():", result.title())
print("swapcase():", result.swapcase())
print("capitalize():", result.capitalize())
print("casefold():", result.casefold())
print("center():", result.center(30))
print("count('a'):", result.count('a'))
print("endswith('a'):", result.endswith('a'))
print("find('a'):", result.find('a'))
print("isalnum():", result.isalnum())
print("isdigit():", result.isdigit())
print("isnumeric():", result.isnumeric())
print("isspace():", result.isspace())
print("replace():", result.replace('a', '@'))
```

```python
a = 10

print("Initial value:", a)

a += 5
print("After += :", a)

a -= 3
print("After -= :", a)

a *= 2
print("After *= :", a)

a /= 4
print("After /= :", a)

a %= 3
print("After %= :", a)

a **= 2
print("After **= :", a)

a //= 2
print("After //= :", a)
```

```python
name = input("Enter student name: ")
student_class = input("Enter class: ")

mark1 = float(input("Enter marks for Subject 1: "))
mark2 = float(input("Enter marks for Subject 2: "))
mark3 = float(input("Enter marks for Subject 3: "))
mark4 = float(input("Enter marks for Subject 4: "))
mark5 = float(input("Enter marks for Subject 5: "))

total = mark1 + mark2 + mark3 + mark4 + mark5
percentage = (total / 500) * 100

if percentage >= 60:
    grade = 'A'
elif percentage >= 50:
    grade = 'B'
elif percentage >= 40:
    grade = 'C'
elif percentage >= 33:
    grade = 'D'
else:
    grade = 'Fail'

print("\n----- Student Result -----")
print("Name:", name)
print("Class:", student_class)
print("Total Marks:", total)
print("Percentage:", percentage, "%")
print("Grade:", grade)
```
