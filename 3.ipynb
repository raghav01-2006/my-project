```python
# Assignment 4

# 1. Function to find maximum of three numbers

def maximum(a, b, c):
    return max(a, b, c)

print("Maximum Number:", maximum(10, 25, 15))


# 2. Function to return distinct elements from a list

def distinct_list(lst):
    new_list = []
    
    for item in lst:
        if item not in new_list:
            new_list.append(item)
    
    return new_list

print("Distinct Elements:", distinct_list([1, 2, 2, 3, 4, 4, 5]))


# 3. Function to multiply all numbers in a list

def multiply_list(lst):
    result = 1
    
    for num in lst:
        result = result * num
    
    return result

print("Multiplication Result:", multiply_list([2, 3, 4]))


# 4. Function to calculate factorial

def factorial(n):
    fact = 1
    
    for i in range(1, n + 1):
        fact = fact * i
    
    return fact

print("Factorial:", factorial(5))


# 5. Program to reverse a string

def reverse_string(text):
    return text[::-1]

print("Reversed String:", reverse_string("Python"))


# 6. Function to check whether number falls within a range

def check_range(num, start, end):
    if num in range(start, end + 1):
        print(num, "is within the range")
    else:
        print(num, "is not within the range")

check_range(7, 1, 10)


# 7. Function to print even numbers from a list

def even_numbers(lst):
    print("Even Numbers:")
    
    for num in lst:
        if num % 2 == 0:
            print(num)

even_numbers([1, 2, 3, 4, 5, 6, 7, 8])


# 8. Function to check prime number

def prime_number(n):
    if n <= 1:
        return False
    
    for i in range(2, n):
        if n % i == 0:
            return False
    
    return True

num = 11

if prime_number(num):
    print(num, "is Prime")
else:
    print(num, "is Not Prime")


# 9. Function to count upper and lower case letters

def count_case(text):
    upper = 0
    lower = 0
    
    for char in text:
        if char.isupper():
            upper += 1
        elif char.islower():
            lower += 1
    
    print("Upper Case Letters:", upper)
    print("Lower Case Letters:", lower)

count_case("Hello Python")


# File Handling Practice

# Writing to a file
file = open("sample.txt", "w")
file.write("Hello, this is file handling in Python.\n")
file.close()

# Reading from a file
file = open("sample.txt", "r")
print("\nFile Content:")
print(file.read())
file.close()

# Appending data to a file
file = open("sample.txt", "a")
file.write("This line is added using append mode.\n")
file.close()

# Reading again after append
file = open("sample.txt", "r")
print("\nUpdated File Content:")
print(file.read())
file.close()
```
